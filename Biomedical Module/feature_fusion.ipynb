```python
import pandas as pd
```

```python
ppg_df = pd.read_csv("../Data/processed/ppg_features.csv")
acc_df = pd.read_csv("../Data/processed/acc_features.csv")

print("PPG:", ppg_df.shape)
print("ACC:", acc_df.shape)
```

```python
ppg_df[["subject_id", "activity"]].head(), acc_df[["subject_id", "activity"]].head()
```

```python
#droping duplicate features
ppg_feat_only = ppg_df.drop(columns=["subject_id", "activity"])
```

```python
fusion_df = pd.concat(
    [acc_df[["subject_id", "activity"]], ppg_feat_only, acc_df.drop(columns=["subject_id", "activity"])],
    axis=1
)

print("Fused shape:", fusion_df.shape)
fusion_df.head()
```

```python
fusion_df.columns
```

```python
OUTPUT_PATH = "../Data/processed/fused_ppg_acc_features.csv"
fusion_df.to_csv(OUTPUT_PATH, index=False)

print("Fused feature dataset saved.")
```
