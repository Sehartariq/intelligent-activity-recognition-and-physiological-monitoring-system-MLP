```python
import os
import pandas as pd
```

```python


DATA_PATH = "../data/raw" # <-- keep your correct path
```

```python
all_data = []
```

```python
for subject_folder in os.listdir(DATA_PATH):
    subject_path = os.path.join(DATA_PATH, subject_folder)

    if os.path.isdir(subject_path):
        for filename in os.listdir(subject_path):
            if filename.endswith(".csv"):

                # ---- Extract subject & activity safely ----
                name = filename.replace(".csv", "")

                if "_" in name:
                    subject_id, activity = name.split("_")
                elif "." in name:
                    subject_id, activity = name.split(".")
                else:
                    continue  # skip weird files

                # ---- Load CSV ----
                file_path = os.path.join(subject_path, filename)
                df = pd.read_csv(file_path)

                # ---- ADD LABELS (THIS WAS THE BUG) ----
                df["subject_id"] = subject_id
                df["activity"] = activity

                all_data.append(df)

```

```python
df.columns.tolist()
```

```python
# ---- COMBINE EVERYTHING ----
combined_df = pd.concat(all_data, ignore_index=True)

print("Shape:", combined_df.shape)
print("Unique subjects:", combined_df["subject_id"].nunique())
print("Activities:", combined_df["activity"].unique())

combined_df.tail()
```
